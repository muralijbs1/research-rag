```python
from IPython.display import Markdown, display

pdf_path = ROOT/"uploaded_pdfs"/"Attention_is _all_you_need.pdf"

chunk_sizes = [200, 300, 500, 750, 1000]
overlaps =    [20,  30,  50,  50,  50]

for size, overlap in zip(chunk_sizes, overlaps):
    chunks = load_and_chunk_pdf(pdf_path, chunk_size=size, overlap=overlap)
    print(f"\n=== chunk_size = {size}, overlap = {overlap} ===")
    print(f"Total chunks: {len(chunks)}")

    num_to_show = min(3, len(chunks))
    for i in range(num_to_show):
        print(f"\n--- Chunk {i+1} (size={size}, overlap={overlap}) ---")
        display(Markdown(chunks[i]))
```
